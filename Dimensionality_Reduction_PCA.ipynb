A major aspect of Machine learning is Data Analysis. The performance of a model depends on the data that is used to train and test it. But working with high dimensional (muliple features) data makes it difficult to analyse or intepret or even visualise the given data[1]. Dimensionality reduction deals with transforming data from a high-dimensional space into a low-dimensional space so that the low-dimensional representation retains some meaningful properties of the original data[2]. Principal Component Analysis (PCA) is a linear dimensionality reduction approach ,based on linear algebra, that has been around for more than a 100 years and is considered as the most common technique for data compression and visualization [1]. In simple words, PCA is a method for compressing a lot of data into something that captures the essence of the original data. It takes a dataset with a lot of dimension (i.e., datasets with a lot of features) and flattens it into 2 or 3 dimensions hence making it easier to analyse the data[3]. # Part 1: Mean and Covariance
In order to understand covariance matrix of a given data, we must understand mean and variance.

Mean is the average data point in a collection of data points.
For example, consider a dataset D= {$x_{1}$, $x_{2}$, $x_{3}$, ...$x_{N}$}.
The mean is given by,

\begin{align}
       	E[d] = \mu =  \frac{1}{N} \sum_{i=1}^N x_{i}  .
    \end{align}

Variance describes the concentration of data points around the mean value, in a given data set.
For the same dataset D,

\begin{align}
        Var[x] = \frac{1}{N} \sum_{i=1}^N (x_{i}- 	\mu)^2  .
    \end{align}

For a single dimension data, the above explanation would suffice, but for higher dimension data, we would need to define the covariance of the data. For a two-dimensional data,

\begin{align}
        cov[x,y] = E[(x-\mu_x)(y-\mu_y)]
        = \frac{1}{N} \sum_{i=1}^N (x_{i}- \mu_x)(y_{i}- \mu_y)
         .
    \end{align}

 We would also need to understand the effects of linear transformation on means and variances of a dataset.

 For mean,

\begin{align}
        E[D\propto]  = \propto E[D] , where \propto = scaling factor \\
        E[D\propto + a] = a + \propto E[D] ,   \ a = shifting factor
         .
    \end{align}

For covariance,
\begin{align}
        var[D]  = var[D+a],   \ a = shifting factor \\
        var[D\propto] = \propto^2 var[D] \\
        var [AD + b] = A.var[D].A^T
         .
    \end{align}
  where A is transformation matrix

# Part 2: Orthogonality and projections

Orthogonality and projections have already been covered in previous modules (Sec 2.6, 2.7 and 3 [1]) , but here is the quickly recap of the main ideas:

When inner product of two vectors is the dot product:
\begin{align}
       	X^TY = \mu =  \sum_{i=1}^N X_{i}.Y_{i}  
    \end{align}

Norm or length of a vector X is:
\begin{align}
       	||X|| = \sqrt(X^T.X) \\
         = \sqrt(\sum_{i=1}^N (X_{i})^2 )     \end{align}

Distance between two vectors X and Y is given by:
\begin{align}
       	d(X,Y) = ||X-Y|| = \sqrt(X-Y)^T.(X-Y)
    \end{align}

Angle between two vectors is given by:
\begin{align}
       	cos\propto = \frac{X^T.Y}{||X||||Y||}
    \end{align}

Project of vector X on subspace U with basic vector b is given as:
\begin{align}
       	\Pi_U(X)
    \end{align}
where,
\begin{align}
       	\Pi_U(X) \in U ⇒ \exists \lambda \in R \\
         \Pi_U(X) = \lambda b
    \end{align}
here λ is the coordinate of the projection and b is the basis vector of subspace U
\begin{align}
       <b,(\Pi_U(X)-X)> =0  
    \end{align}
which is the orthogonality condition

so
\begin{align}
       <b,(\Pi_U(X)-X)> =0 \\
       ⇒<b,\Pi_U(X)> - <b,X> = 0\\
       ⇒ <b, \lambda b> - <b,X> = 0 \\
       ⇒ \lambda<b,b> - <b,X> = 0 \\
       ⇒ \lambda = \frac{<b,X>}{||b||^2} \\
       \therefore \Pi_U(X) = \lambda b = \frac{<b,X>}{||b||^2}.b
    \end{align}

For projections onto higher dimensional space, where subspace U is defined by basis vectors [$b_1$, $b_2$] and X ∈ $R^3$,
\begin{align}
       \Pi_U(X) = \lambda_1 b_1 + \lambda_2 b_2  .
    \end{align}
additionally,
\begin{align}
       <X-\Pi_U(X), b_1> = 0 \\
       <X-\Pi_U(X), b_2> = 0  .
    \end{align}


#Part 3: Eigen Values and eigen vectors

Eigen values and eigen vectors have been covered in detail in section 4.2 [1].
To recap, eigen vectors are those vectors that remain on the same span of the original vector after undergoing a transformation. The factor to which these vectors shrink or expand is known as the eigen value.
Eigen values can be found using:
\begin{align}
       A - \lambda I = 0
    \end{align}
where A is the transformation matrix.

To find the eigen vectors, just replace the λ with eigen values in the following equation and solve:
\begin{align}
       (A - \lambda I)x = 0
    \end{align}
Do remember that every eigenvector has an eigenvalue. Also, the number of eigen vectors is equal to the number of dimensions of the data. For example, for a 3-dimensional dataset, there are 3 variables, therefore there are 3 eigenvectors with 3 corresponding eigenvalues.

Now, why are eigen values and eigen vectors important. First let's understand principal components.

Principal components are variables that are derived from linear combination of existing variables. The combination is done in such a way that the resulting principal components are uncorrelated and the first component holds the maximum amount of information of the initial data, followed by the next component and so on. In simple terms, if there is a 10 dimensional data, it can have 10 principal components where the first principal component can have the maximum information of the initial data, followed by the second principal component and so on and so forth [4].

Geometrically speaking, principal components represent the directions of the data that explain a maximal amount of variance. The idea is, larger the variance of data carried by a line, the larger the dispersion of the data points along it, and the larger the dispersion along a line, the more information it has. In other words, principal components become the new axes that provide the best angle to analyse the differnce between the data.

What do principal components have to do with eigen values and eigen vectors? **Everything**

Eigen vectors of a covariance matrix are actually the directions of the axes where there is the most variance(most information). In other words, they are the Principal Components. Eigenvalues are simply the coefficients attached to eigenvectors, which give the amount of variance carried in each Principal Component.

Hence the eigen vector having the highest eigen value is the first principal component. The second highest eigen value's vector becomes the second principal component.

In short, by ranking the eigenvectors in order of their eigenvalues, highest to lowest, you get the principal components in order of significance.



# Part 4: The PCA calculation

Assume we have a centered dataset $x_1$, . . . , $x_N$ , $x_n$ ∈ $R^D$ (i.e., E[$x_n$]=0) . Then the data covariance matrix is given as:

\begin{align}
       	S= \frac{1}{N}XX^T  \in R^{D*D} ... (1)
    \end{align}

where X = [x1, . . . , xN ] is a D × N matrix whose columns are the data
points.

Now assume that N << D, i.e., the number of data points is smaller
than the dimensionality of the data. If there are no duplicate data points,
the rank of the covariance matrix S is N, which would also be the number of eigen values for S, calculated by:
\begin{align}
       	Sb_m= \lambda_m*b_m ...(2)
    \end{align}
where m = 1, . . . , M and $b_m$ is the basic vector of the principal subspace

Substituing (1) in (2), we get:
\begin{align}
       	\frac{1}{N}XX^Tb_m= \lambda_m*b_m
    \end{align}

 On multiplying  $X^T$ ∈ $R^{NXD}$ from the left-hand side, we get:
 \begin{align}
       	\frac{1}{N}X^TXX^Tb_m= \lambda_m*X^Tb_m \\
         ⇒\frac{1}{N}X^TXc_m= \lambda_m*c_m
    \end{align}
we get a new eigenvector/eigenvalue equation: $λ_m$ remains eigen-
value and the eigen vector is given by $c_m$ = $X^T$$b_m$

In order to calculate the projection of data points x ∈ $R^D$ on principal subspace,
first we standardise the data points:
\begin{align}
       	x^d = \frac{x^d - \mu^d}{std.dev^d} \\
    \end{align}

where d = d = 1, . . . , D and $x^d$ is dth component of x

The projection is calculated as:
\begin{align}
        \tilde{x} = BB^Tx \\
    \end{align}
where $z$ = $B^T$$x$ is the code or coordinates
Here, B is the matrix that contains the eigenvectors that are associated with the largest eigenvalues of the data covariance matrix as columns.
Please note that PCA returns the coordinates, not the projections $x$.

#Let's put it all together

The basic algorithm of PCA is:
1. STANDARDIZATION : This step involves subtracting the mean from the data points and then dividing by standard deviation (refer part 1). This step is required to standardise the initial variables to comparable scales so that each variable can contribute equally to the analysis.

\begin{align}
       	z = \mu =  \frac{value-\mu}{std. dev}
    \end{align}

2. COVARIANCE MATRIX COMPUTATION: covariance matrix is a p × p symmetric matrix (where p is the number of dimensions) that has covariances associated with all possible pairs of the initial variables. This is done to identify correlation between redundant features. For a 3-dimensional dataset (x,y,z), the covariance matrix is:

\begin{align}
        \begin{pmatrix}
        Cov(x,x)=Var(x) & Cov(x,y) & Cov(x,z) \\
        Cov(y,x) & Cov(y,y)=Var(y) & Cov(y,z) \\
        Cov(z,x) & Cov(z,y) & Cov(z,z)=Var(z) \\
        \end{pmatrix}.
    \end{align}

If covariance value is:
* negative: two values are inversely correlated
* positive: two values are correlated

3. COMPUTE THE EIGENVECTORS AND EIGENVALUES: The eigenvalues and eigenvectors of the covariandce matrix is calculated in order to find the principal components of this matrix. We can find the first and subsequent principal components by ranking the eigen vectors based on their  eigen values, from highest to the lowest.

4. FINALISE THE FEATURE VECTOR: In this step, we basically choose the first and second (and some cases the third) principal components and discard the others and form the feature vector. In other words, the feature vector is a matrix that contains the shortlisted eigenvectors having highest eigen values, as columns.

5. RECAST THE DATA ALONG THE PRINCIPAL COMPONENTS AXES: The aim is to use the feature vector formed using the eigenvectors of the covariance matrix, to reorient the data from the original axes to the ones represented by the principal components (hence the name Principal Components Analysis). The resultant dataset is obtained as:

\begin{align}
       	FinalDataSet =  FeatureVector^T * StandardisedOriginalDataSet^T
    \end{align}

```python
#let's initialise the libraries that we would call:
import numpy as np # linear algebra
import pandas as pd # data processing

from google.colab import drive
drive.mount("/content/drive")
```

```python
import matplotlib.pyplot as plt
dataset =pd.read_csv('/content/drive/MyDrive/input/mnist_train.csv')

print(dataset.head(5))
```

Notice that the first column, label, is the actual digit and the remaining 784 columns is data in matrix format of that digit. Let's save the label in a separate label and drop it from the original dataset.

```python
# save the labels into a variable named label
labels = dataset['label']

# drop the label feature and store the pixel data in d
data = dataset.drop('label', axis=1)
```

```python
print(data.shape)
print(labels.shape)
```

The training set has about 60000 images of handwritten digits. To view the digit, execute the code given below

```python
# display or plot a number
plt.figure(figsize=(7,7))
index=165 #taking any random row index of the dataset

# reshape from 1d(1x784) to 2d (28x28) pixel array
grid_data = data.iloc[index].values.reshape(28,28)
plt.imshow(grid_data, interpolation='none', cmap='gray')
plt.show()

# print the label of the corresponing index value from the 'label' dataset
print(labels[index])
```

Let's start the PCA process to visualise the data using method 1: step by step alogorithm of PCA


```python
# Step 1: Standardizing the data
# standardizing means calculating the z score
from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(data)
print(standardized_data.shape)
```

```python
# Step 2: find the covariance matrix which is : A^T * A
sample_data = standardized_data

#matrix multiplication using numpy
covar_matrix = np.matmul(sample_data.T, sample_data)
print("The shape of variance matrix = ", covar_matrix.shape)
```

```python
# Step 3 & 4: finding the top two eigen-values and corresponding eigen-vectors
# and creating feature vector of two principal components
from scipy.linalg import eigh

# the parameter 'eigvals' is defined (low value to high value)
# eigh function will return the eigen values in ascending order
# this code generates only the top 2 (783 and 782) eigenvalues.
values, vectors = eigh(covar_matrix, eigvals=(782, 783))

print("Shape of eigen vectors = ", vectors.shape)

# converting the eigen vectors into 2d shape for easyness of further computation
vectors = vectors.T

print("Updated shape of eigen vectors = ", vectors.shape)
# here the vectors[1] represent the eigen vector corresponding 1st principal eigen vector
# here the vectors[0] represent the eigen vector corresponding 2nd principal eigen vector

#swapping rows of vectors matrix and values
vectors[[0,1]]= vectors[[1,0]]
values[[0,1]] = values[[1,0]]

# now the vectors[0] (and values[0]) represent the eigen vector corresponding 1st principal eigen vector
# now the vectors[1] (and values[1]) represent the eigen vector corresponding 2nd principal eigen vector
```

```python
# To see the eigen values from highest to lowest
values
```

```python
#to see the eigen vectors based on higest to lowest eigen values
vectors
```

```python
# Step 5:projecting the original data sample on the plane
# formed by two principal eigen vectors by vector-vector mulitplication
new_coordinates = np.matmul(vectors, sample_data.T)
print ("Resultant new data points' shape ", vectors.shape, "X", sample_data.T.shape," = ", new_coordinates.shape)
```

```python
# appending label to the 2nd projected data
new_coordinates = np.vstack((new_coordinates, labels)).T

# creating a new data frame for plotting the labeled points
dataframe = pd.DataFrame(data=new_coordinates, columns=('1st_principal', '2nd_principal', 'label'))
print(dataframe.head())
```

```python
# finally, ploting the 2d data points with seaborn
import seaborn as sn
sn.FacetGrid(dataframe, hue="label", height=8).map(plt.scatter, '1st_principal', '2nd_principal').add_legend()
plt.show()
```

Now let's compute the PCA through method 2: utilising the inbuilt PCA function of Scikit-learn library

```python
from sklearn.decomposition import PCA
# initializing the pca
pca = PCA(n_components=2) # configuring the number of components
pca_data = pca.fit_transform(sample_data)
```

```python
print("Shape of pca_reduced.shape = ", pca_data.shape)
```

```python
# attaching the label for each 2-d data point
pca_data = np.vstack((pca_data.T, labels)).T

# creating a new data frame which help us in ploting the result data
pca_df = pd.DataFrame(data=pca_data, columns=('1st_principal', '2nd_principal', 'label'))
sn.FacetGrid(pca_df, hue='label', height=6).map(plt.scatter, '1st_principal', '2nd_principal').add_legend()
plt.show()
```
